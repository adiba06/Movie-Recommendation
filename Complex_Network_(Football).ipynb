```
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def average_degree(g):
    tot_deg = 0
    for node in g.nodes:
        tot_deg += g.degree(node)

    return tot_deg / g.number_of_nodes()
```

```
def degree_structure(g):
    mp = {}
    for node in g.nodes:
        deg = g.degree(node)
        if deg not in mp:
            mp[deg] = 0

        mp[deg] += 1

    print("Degree".ljust(10), "Number")
    x = []
    y = []
    for key in mp:
        val = mp[key]
        x.append(key)
        y.append(val)
        print(str(key).ljust(10), val)

    return x, y
```

```
# load the graph
g = nx.read_gml('./football.gml')
```

```
print('Number of nodes:', g.number_of_nodes())
```

```
print('Nodes are')
for node in g.nodes:
    print(node)
```

We can see that the average degree of this graph is nearly 10. It means that every node is connected with 10 nodes on average. 

The max possible degree for this graph can be 114. As the total number of nodes in the graph is 115, so one node cna be connected to 114 nodes at most.

```
print('Number of edges:', g.number_of_edges())
print(average_degree(g))
```

From the degree structure, we can see that, most of the nodes (66) are connected to 11 nodes, and on the second place, 28 nodes are connected with 10 nodes.

```
x, y = degree_structure(g)
```

```
plt.plot(x, y)
plt.xlabel('Degree')
plt.ylabel('Number of Nodes')
plt.show()
```
